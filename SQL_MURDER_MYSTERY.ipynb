```python
import pandas as pd
import sqlite3 as sql

#Connecting database
connection=sql.connect("/sql-murder-mystery.db")

#create cursor object, required to fetch results FROM SQL queries
mycursor = connection.cursor() #Cursor function to connect with table
pd.set_option('display.max_colwidth', None)
```

#Retriving Table names

```python
query="""SELECT name FROM sqlite_master
WHERE type='table';""" #SQL Query to extract data
print("TABLES IN THE DATABASE\n")
mycursor.execute(query) #Executing Query
read_data=pd.read_sql_query(query,connection) #to fetch data FROM table to show
read_data
```

```python
print("\nCRIME SCENE REPORT\n")
query="""SELECT * FROM crime_scene_report where date=20180115 AND
type="murder" AND
city="SQL City";
""" #SQL Query to extract read_data
mycursor.execute(query) #Executing Query
read_data=mycursor.fetchall()
print("Data of murder on 15 January 2018 in SQL City \n")
for Record_no, records in enumerate(read_data, start=1):
    print(f"{Record_no}",read_data)
```
